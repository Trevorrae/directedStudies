```python
def betterOrWorse(x,y,healthierOrSicker,coefficent):
    labrange = x-y
    temp = round(random.uniform(labrange,1),2)
    if betterOrWorse == 1:
        result = round((x-(temp*coefficent)),2)
    else:
        result = round((y+(temp*coefficent)),2)
    return result
```
